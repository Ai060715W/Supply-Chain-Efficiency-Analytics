```python
import os
import pickle
import pandas as pd
import plotly.express as px
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

PROJECT_ROOT = r'D:\Supply-Chain-Efficiency-Analytics'
DATA_DIR = os.path.join(PROJECT_ROOT,'data','raw')
DASHBOARD_DIR = os.path.join(PROJECT_ROOT,'dashboards')

sales_orders = pd.read_csv(os.path.join(DATA_DIR,'sales_orders.csv'))
sales_orders['sales_order_date'] = pd.to_datetime(sales_orders['sales_order_date'])

# 简单线性回归预测销售量（按产品汇总月度）
monthly_sales = sales_orders.groupby([pd.Grouper(key='sales_order_date', freq='M')])['quantity_sold'].sum().reset_index()
monthly_sales['month_num'] = range(len(monthly_sales))
X = monthly_sales[['month_num']]
y = monthly_sales['quantity_sold']

model = LinearRegression()
model.fit(X, y)
monthly_sales['forecast'] = model.predict(X)

fig_sales_forecast = px.line(monthly_sales, x='sales_order_date', y=['quantity_sold','forecast'], title='销售预测')

# 保存 Pickle
modeling_results = {
    'fig_sales_forecast': fig_sales_forecast
}

with open(os.path.join(DASHBOARD_DIR,'modeling_results.pkl'),'wb') as f:
    pickle.dump(modeling_results, f)

print("[OK] Modeling Pickle 已生成")
```
